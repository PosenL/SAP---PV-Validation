```python
class Email_generate:
    def __init__(self):
        self.today_m_d = datetime.today().strftime("%m/%d")
        self.recipients = pd.read_excel(r".\Input\Materials\Email Setup.xlsx")
        self.work_week = 'WK' + str(datetime.today().year)[-2:] + str(datetime.today().isocalendar().week)
        self.hyperlink_date_lvl = 'https://nxp1.sharepoint.com/teams/14_5_6/GMDM%20Documents/Forms/AllItems.aspx?id=%2Fteams%2F14%5F5%5F6%2FGMDM%20Documents%2FEM%20PDM%2FActivities%2FPV%20Validation%20Tool&viewid=c4f82d43%2Dd8b2%2D4dbb%2D927f%2D08d151cdd39e'
        self.hyperlink_display_text_date_lvl = 'EMPDM Folder'
        self.hyperlink = 'https://nxp1.sharepoint.com/teams/14_5_6/GMDM%20Documents/Forms/AllItems.aspx?id=%2Fteams%2F14%5F5%5F6%2FGMDM%20Documents%2FEM%20PDM%2FActivities%2FPV%20Validation%20Tool&viewid=c4f82d43%2Dd8b2%2D4dbb%2D927f%2D08d151cdd39e'
        self.hyperlink_display_text = 'Global Master Data Management - PV Validation Tool - All Documents (sharepoint.com)'


    def Generate_pivot(self, input_export_summary_table):
        df_pivot = pd.DataFrame(input_export_summary_table.groupby('Mismatched_Columns').size().values, 
                                index = input_export_summary_table.groupby('Mismatched_Columns').size().index,
                                columns = ["Count of mismatched PVs"])

        def add_empty_rows(df_input):
            # Create a new DataFrame with the same columns
            df_input["Mismatched Combinations"] = df_input.index
            df_input.index = range(len(df_input))
            empty_df = pd.DataFrame(columns=df_input.columns)
        
            # Iterate over the rows in the original DataFrame
            for idx in range(len(df_input)):
                # Append the original row
                empty_df = empty_df.append(df_input.iloc[idx])
                # Append an empty row
                empty_df = empty_df.append(pd.Series(), ignore_index=True)
            return empty_df


        df_pivot = add_empty_rows(df_pivot)
        num_of_mismatch = df_pivot["Count of mismatched PVs"].sum()
        df_pivot = df_pivot.append(pd.DataFrame({"Mismatched Combinations":["Grand Total"], "Count of mismatched PVs":[num_of_mismatch]}))
        df_pivot = df_pivot[["Mismatched Combinations", "Count of mismatched PVs"]]
        return df_pivot
    


    def table_style(self, wording_html_type):
        if wording_html_type == "th":
            th_style = '''
                <th style = "background-color: #305496;
                color:white;
                font-family: Calibri;
                font-size: 13;
                text-align: left;
                border-style: solid; 
                border: 1px solid black; 
                ">
                '''
            return th_style
        
        if wording_html_type == "td":
            td_style = '''
            <td style = "
            font-family: Calibri;
            font-size: 13;
            text-align: left;
            border-style: solid; 
            border: 1px solid black;
            "> 
            '''
            return td_style

    def Email_wording(self, input_analyzed_frame):
        if len(input_analyzed_frame) != 0:
            html_body = f"""
                <html>
                <head>
                </head>

                <body>
                    <p>
                    <b>Hi EMPDM Team, </b>
                    <br/>The summary table of the <u>{self.today_m_d}</u> PV Validation run is below. All values have been updated via LCO/C223.<br/>
                    Therefore, all the active PVs in SAP are maintained with correct values.<br/> 
                    Please find the details in folder: <a href='{self.hyperlink_date_lvl}'> 
                    {self.hyperlink_display_text_date_lvl}</a>.<br/>
                    Please also find the : <a href='{self.hyperlink}'>
                    {self.hyperlink_display_text}</a>.<br/> 

                    <br>
                    <b><u>Process Time:</u></b><br/>
                    {self.export_process_time_table}
                    
                    <br>
                    <b><u>Summary Table:</u></b>
                    <br/>


                    <br/>
                    <li> PV Details Summary </li> 
                    {self.export_detail_summary_table}
                    
                    <br/>
                    <li> Mismatched Attributes Summary </li> 
                    {self.export_mismatched_attributes_table}

                    <br/>
                    <li> Summary table of each error type combination</li>
                    {self.export_summary_error_type_table}
                    </p>
                </body>

                </html>
                """.format(
                            self.today_m_d,
                            self.hyperlink_date_lvl, self.hyperlink_display_text_date_lvl,
                            self.hyperlink, self.hyperlink_display_text,
                            self.export_process_time_table,
                            self.export_detail_summary_table,
                            self.export_mismatched_attributes_table,
                            self.export_summary_error_type_table
                            )

        else:
            html_body = f"""
                <html>
                <head>
                </head>

                <body>
                    <p>
                    <b>Hi EMPDM Team, </b>
                    <br/>The summary table of the MM/DD PV Validation run is below. No error/mismatch was found today.<br/>
                    Therefore, all the active PVs in SAP are maintained with correct values.<br/> 
                    Please find the details in folder: <a href='{self.hyperlink_date_lvl}'> 
                    {self.hyperlink_display_text_date_lvl}</a>.<br/>
                    Please also find the : <a href='{self.hyperlink}'>
                    {self.hyperlink_display_text}</a>.<br/> 

                    <br>
                    <b><u>Process Time:</u></b><br/>
                    {self.export_process_time_table}
                    
                    <br>
                    <b><u>Summary Table:</u></b>
                    <br/>


                    <br/>
                    <li> PV Details Summary </li> 
                    {self.export_detail_summary_table}
                    
                    <br/>
                    <li> Mismatched Attributes Summary </li> 
                    {self.export_mismatched_attributes_table}

                    </p>
                </body>

                </html>
                """.format(
                            self.today_m_d,
                            self.hyperlink_date_lvl, self.hyperlink_display_text_date_lvl,
                            self.hyperlink, self.hyperlink_display_text,
                            self.export_process_time_table,
                            self.export_detail_summary_table,
                            self.export_mismatched_attributes_table
                            )
        return html_body

        
    def Email_setup(self, export_detail_summary_table, export_mismatched_attributes_table, export_summary_error_type_table):
        outlook = win32com.client.Dispatch('Outlook.Application')
        message = outlook.CreateItem(0)
        message.Display()

        send_list = self.recipients["Send to"].dropna().tolist()
        cc_list = self.recipients["CC to"].dropna().tolist()

        message.To = ';'.join(send_list)
        message.Cc = ';'.join(cc_list)
        message.Subject = "PV Validation Tool Run Result - {} ({})".format(self.work_week, self.today_m_d)
        

        empty_process_time_table  = pd.DataFrame([["Human TPT", ""], ["Robot TPT", ""]], columns= ["Processing Time", str(datetime.today().strftime("%Y/%m/%d"))])
        self.export_process_time_table = empty_process_time_table.to_html(index = False, border=True).replace('<th>', self.table_style('th')).replace('<td>', self.table_style('td')).replace('#305496', '#5F9EA0').replace("color:white","color:black")
        self.export_detail_summary_table = export_detail_summary_table.to_html(index = True, border=True).replace('<th>', self.table_style('th')).replace('<td>', self.table_style('td'))
        self.export_mismatched_attributes_table = export_mismatched_attributes_table.to_html(index = True, border=True).replace('<th>', self.table_style('th')).replace('<td>', self.table_style('td'))
        
        if len(export_summary_error_type_table) != 0:
            export_summary_error_type_table = self.Generate_pivot(export_summary_error_type_table)
            export_summary_error_type_table.fillna('', inplace=True)
            self.export_summary_error_type_table =  build_table(export_summary_error_type_table, 'blue_dark', font_size = 13, font_family='Calibri', index=False)
        
        message.HTMLBody = self.Email_wording(export_summary_error_type_table)
        message.Display()
    pass
```

```python
class Data_Preprocessing:
    def __init__(self):
        self.today = (datetime.today()).strftime('%Y%m%d')
        self.output_folder = "./Output/" + self.today
        self.ZQRYFI = pd.read_excel(".\\Input\\ZQRYFI.xlsx").copy()
        self.ZQRYFI_orig = pd.read_excel(".\\Input\\ZQRYFI.xlsx").copy()
        self.ZECMM772_orig = pd.read_excel(".\\Input\\ZECMM772.xlsx").copy()
        self.ZECMM772, self.Invalid_ZECMM772 = self.ZECMM772_filter_out_duplicate(self.ZECMM772_orig)
        self.tblRequestedtoCOE = pd.read_excel(".\\Input\\tblRequestedtoCOE.xlsx").copy()
        self.PLC_PIR_per_wave = pd.read_excel(".\\Input\\PLC PIR per wave_consolidated.xlsx").copy()
        self.Source_Process_Text = pd.read_excel(".\\Input\\Source Value-StAr Processsing.xlsx", sheet_name="Process Text Per E-Plant").copy()
        self.Source_COEL_LORE = pd.read_excel(".\\Input\\Source Value-StAr Processsing.xlsx", sheet_name="EM Guide - COEL_LORE").copy()
        self.Source_file_plant_naming_mapping = dict(zip(self.Source_Process_Text.drop_duplicates("E-Plant")["E-Plant"], 
                                                         self.Source_Process_Text.drop_duplicates("E-Plant")["Naming Rule"]))
        
        REP077_available_day = glob.glob("./Input/REP-D077_*.xlsx")[0].replace("\\", "/").split("_")[-1].split(".")[0]
        self.REP_D077 = pd.read_excel(".\\Input\\REP-D077_{}.xlsx".format(REP077_available_day), usecols = ["DWFR_Name", "DIFF_Name"])
        self.REP_D077["DWFR_Name"] = self.REP_D077["DWFR_Name"].str.replace(' ', "")
        REP_D077_drop_dup = self.REP_D077.drop_duplicates("DWFR_Name")
        REP_D077_drop_dup_map = dict(zip(REP_D077_drop_dup["DWFR_Name"], REP_D077_drop_dup["DIFF_Name"]))
        self.REP_D077_drop_dup_map = REP_D077_drop_dup_map.copy()


    def if_folder_exist(self, input_path):  # Create the output folder if not exist
        if not os.path.exists(input_path):
            os.makedirs(input_path)
        pass

    def main_consolidate_result(self):
        self.if_folder_exist(self.output_folder)
        self.ZECMM772["Info Short Text"] = self.ZECMM772["Info Short Text"]+ " " + self.ZECMM772["Plant"]
        self.ZQRYFI["PIR Text"] = self.ZQRYFI["PIR Text"]+ " " + self.ZQRYFI["Plnt"]


        self.ZQRYFI_Eplant_without_Naming_Rule =self.main_process_ZQRYFI()
        self.tblRequestedtoCOE_Eplant_without_Naming_Rule = self.main_process_tblRequestedtoCOE()
        self.main_process_PLC_PIR_per_Wave()
        

        ZQRYFI_Consolidate_Result = self.ZQRYFI.copy()

        ZQRYFI_Consolidate_Result = self.report_mapping(ZQRYFI_Consolidate_Result, self.ZECMM772, "PIR Text", "Info Short Text", 'PIR nr', "INNU_ZECMM_PIRTEXT", "Info Rec.", "MATCH_INNU_PIRTEXT", FALSE)
        ZQRYFI_Consolidate_Result = self.report_mapping(ZQRYFI_Consolidate_Result, self.ZECMM772, "ID_KEY", "Info Short Text", 'PIR nr', "INNU_ZECMM_IDKEY", "Info Rec.", "MATCH_INNU_IDKEY", FALSE)
        ZQRYFI_Consolidate_Result = self.report_mapping(ZQRYFI_Consolidate_Result, self.Source_COEL_LORE, "Process Text", "Recipe Operation", 'Logistics Resource/', "Work_Center_SOURCE", "Work Center", "MATCH_WorkCenter", FALSE)
        ZQRYFI_Consolidate_Result = self.report_mapping(ZQRYFI_Consolidate_Result, self.Source_COEL_LORE, "Process Text", "Recipe Operation", 'Cost Element', "Cost Elem_SOURCE", "Cost Elem.", "MATCH_CostElem", FALSE)
        ZQRYFI_Consolidate_Result = self.report_mapping(ZQRYFI_Consolidate_Result, self.tblRequestedtoCOE, "ID_KEY", "ID_KEY", "OANumber_AGRE", "Outline agreement_PIRConsolidation_AGRE", "Outline agreement", "MATCH_AGRE", FALSE)
        ZQRYFI_Consolidate_Result = self.report_mapping(ZQRYFI_Consolidate_Result, self.tblRequestedtoCOE, "ID_KEY", "ID_KEY", "OAItemNumber_AGRP", "Item outline agreement_PIRConsolidation_AGRP", "Item outline agreement", "MATCH_AGRP", FALSE)
        ZQRYFI_Consolidate_Result = self.report_mapping(ZQRYFI_Consolidate_Result, self.PLC_PIR_per_wave, "ID_KEY", "ID_KEY", "NEW OA", "Outline agreement_PIRConsolidation_AGRE", "Outline agreement", "MATCH_AGRE", True)
        ZQRYFI_Consolidate_Result = self.report_mapping(ZQRYFI_Consolidate_Result, self.PLC_PIR_per_wave, "ID_KEY", "ID_KEY", "NEW ITEM", "Item outline agreement_PIRConsolidation_AGRP", "Item outline agreement", "MATCH_AGRP", True)
        
        mask = ZQRYFI_Consolidate_Result[[x for x in ZQRYFI_Consolidate_Result.columns if "MATCH" in x]].apply(lambda row: self.find_string(row, "N"), axis=1)
        ZQRYFI_Consolidate_Result["Correction Needed"]= "N"
        ZQRYFI_Consolidate_Result.loc[ZQRYFI_Consolidate_Result[mask].index, "Correction Needed"]= "Y"
        ZQRYFI_Consolidate_Result.loc[ZQRYFI_Consolidate_Result["Correction Needed"]=="N"].head()

        cols_to_be_check = [x for x in ZQRYFI_Consolidate_Result.columns if "MATCH" in x]
        ZQRYFI_Consolidate_Result["Mismatched_Columns"] = ZQRYFI_Consolidate_Result[cols_to_be_check].apply(lambda row: self.cols_to_be_correct(row, cols_to_be_check), axis=1)
        ZQRYFI_Consolidate_Result = self.broken_flow_check(ZQRYFI_Consolidate_Result, "Material Number", "Process Text", "Plnt")

        #Addtional condition for those PIRTEXT mismatched but IDKEY matched
        ZQRYFI_Consolidate_Result.loc[(ZQRYFI_Consolidate_Result["MATCH_INNU_PIRTEXT"] =="N") & (ZQRYFI_Consolidate_Result["MATCH_INNU_IDKEY"]=="Y"), "MATCH_INNU_PIRTEXT"] ="Skip"
        ZQRYFI_Consolidate_Result.reset_index(drop = True, inplace=True)
        
        Export_File = []
        Export_File.append([self.Invalid_ZECMM772, "Invalid_Duplicate_ZECMM772", len(self.Invalid_ZECMM772)])
        Export_File.append([self.ZQRYFI_Eplant_without_Naming_Rule, "ZQRYFI_Eplant_without_N_Rule", len(self.ZQRYFI_Eplant_without_Naming_Rule)])
        Export_File.append([self.tblRequestedtoCOE_Eplant_without_Naming_Rule, "COE_Eplant_without_N_Rule", len(self.tblRequestedtoCOE_Eplant_without_Naming_Rule)])

        Total_records, Mismatched_PVs = len(ZQRYFI_Consolidate_Result), len(ZQRYFI_Consolidate_Result[ZQRYFI_Consolidate_Result["Correction Needed"] == "Y"])
        Matched_PVs = Total_records - Mismatched_PVs

        
        Overall_Table = pd.DataFrame([Mismatched_PVs, Matched_PVs, Total_records], columns = [(datetime.today()).strftime('%Y/%m/%d')],
                                            index = ["PVs with Mismatch attributes",
                                                    "PVs without Mismatch attributes",
                                                    "Total number of PVs"])
        Overall_Table["Percentage"] = Overall_Table[(datetime.today()).strftime('%Y/%m/%d')].apply(lambda x : "{:.2%}".format(x/Total_records))


        # Export all tables into excel file: ZQRYFI_Consolidate_Result
        with pd.ExcelWriter('./Output/{today_date}/ZQRYFI_Consolidate_Result_{today_date}.xlsx'.format(today_date = self.today)) as writer:
            self.summary_input_process().to_excel(writer, startrow= 0, sheet_name='Summary', index = True)
            Overall_Table.to_excel(writer, startrow= 3, sheet_name='Summary', index = True)
            summary_table = self.summary_table_process(cols_to_be_check, ZQRYFI_Consolidate_Result, Export_File)
            summary_table.to_excel(writer, startrow= 8, sheet_name='Summary', index = True)
            ZQRYFI_Consolidate_Result.to_excel(writer, sheet_name='Detail', index = True)

            df_missing_info= pd.DataFrame([], columns = [x[1] for x in Export_File])
            for sheet in Export_File:
                sheet[0].to_excel(writer, sheet_name=sheet[1], index = False)
                df_missing_info.loc["Total Records", sheet[1]] = sheet[2]
                print("File is now exported with tab: {} !! ".format(sheet[1]))
            df_missing_info.to_excel(writer, startrow= 14, sheet_name='Summary', index = True)


        # Generate Email Template
        ZQRYFI_Consolidate_Result["Mismatched_Columns"] = ZQRYFI_Consolidate_Result["Mismatched_Columns"].astype(str)
        Email_generate().Email_setup(Overall_Table, 
                                     summary_table.fillna(-1).astype(int).replace(-1, ""), 
                                     ZQRYFI_Consolidate_Result[ZQRYFI_Consolidate_Result["Mismatched_Columns"] != '[]'])

        # Message box
        box = tk.Tk()
        box.withdraw()
        messagebox.showinfo('Notification', "Done with ZQRYFI_Consolidate_Result report !!! ")
        os.startfile('{}/Output/{today_date}/ZQRYFI_Consolidate_Result_{today_date}.xlsx'.format(os.getcwd().replace("\\", "/"), today_date = self.today))
        pass 
    
    def summary_input_process(self):
        input_files = {
                    "ZQRYFI": self.ZQRYFI_orig,
                    "ZECMM772": self.ZECMM772,
                    "tblRequestedtoCOE":self.tblRequestedtoCOE,
                    "PLC_PIR_per_wave":self.PLC_PIR_per_wave,
                    "Source_Process_Text":self.Source_Process_Text,
                    "Source_COEL_LORE":self.Source_COEL_LORE,
                    "REP-D077":self.REP_D077
                    }

        allfiles=dict()

        for f in input_files.items():
            allfiles[f[0]] = len(f[1])
        combination = pd.DataFrame.from_dict(allfiles, orient = "index").T
        combination.rename({0:"Total Records"}, inplace=True)     
        return combination
    
    def summary_table_process(self, cols_to_be_check, input_consolidate_file, input_Export_File_list):
        cols_to_be_check = [x for x in input_consolidate_file.columns if "MATCH" in x]
        Combination = pd.DataFrame([])
        for attr in cols_to_be_check:
            Combination = pd.concat([Combination, input_consolidate_file.groupby(attr)[attr].count()], axis = 1)
        Combination.loc["Total Records"] = Combination.sum(axis=0)
        Combination.index = Combination.index.map({"Total Records":"Total Records", "N":"Mismatch", "Y":"Match", "Unmappable": "Unmappable"})
        Combination.rename(columns={"MATCH_INNU_PIRTEXT":"INNU_PIRTEXT",
                                    "MATCH_INNU_IDKEY":"INNU_IDKEY",
                                    "MATCH_WorkCenter":"WorkCenter",
                                    "MATCH_CostElem":"CostElem",
                                    "MATCH_AGRE":"AGRE",
                                    "MATCH_AGRP":"AGRP"}, inplace = True)
        

        Combination = Combination.reindex(columns =["INNU_IDKEY", "INNU_PIRTEXT", "WorkCenter", "CostElem", "AGRE", "AGRP"])

        if "Mismatch" in Combination.index:
            Combination.fillna(0, inplace=True)
            Combination.loc["Mismatch", "Total Mismatched Attributes"] = Combination.loc["Mismatch"].sum()
        else:
            Combination.loc["Mismatch", :] = 0
            Combination.loc["Mismatch", "Total Mismatched Attributes"] = 0


        custom_dict = {"Mismatch":0, "Match":1, "Unmappable":2, "Total Records": 3}
        Combination = Combination.sort_index(key=lambda x: x.map(custom_dict), ascending = True)
        return Combination


    def main_process_ZQRYFI(self):
        diff_exist, diff_not_exist = [], []

        self.ZQRYFI["ID_KEY"] = pd.NaT

        ##Updated self.ZQRYFI["Task list description"] to self.ZQRYFI["Task list description"] on 2024/03/05
        self.ZQRYFI["Process Text"] = self.ZQRYFI["Production Version"].apply(lambda x: self.Process_Text_processing(x))

        #Add new column Naming Rule in ZQRYFI  by naming rule in source file
        self.ZQRYFI["Naming Rule"] = self.ZQRYFI["Plnt"].map(self.Source_file_plant_naming_mapping)

        #Check if there is any Eplant with no naming rule
        ZQRYFI_Eplant_without_Naming_Rule = self.ZQRYFI.loc[self.ZQRYFI["Naming Rule"].isna()].copy()

        if len(self.ZQRYFI.loc[self.ZQRYFI["Naming Rule"].isna()]["Plnt"]) !=0:
            print("ZQRYFI_Eplant without Naming Rule: ", self.ZQRYFI.loc[self.ZQRYFI["Naming Rule"].isna()]["Plnt"].values)
        else:
            print("ZQRYFI_Eplant without Naming Rule: ".format("N/A"))

        self.ZQRYFI = self.ZQRYFI.loc[~self.ZQRYFI["Naming Rule"].isna()]

        self.ZQRYFI["ID_KEY"] = self.ZQRYFI.apply(lambda x : self.NamingRule_processing(x, "Material Number", "BOM Component", "Process Text", "Plnt",
                                                                                 diff_exist, diff_not_exist, self.REP_D077_drop_dup_map), axis=1)

        ZQRYFI_Diff_not_exist = pd.DataFrame(diff_not_exist, columns = ["Consumed", "Produced", "Process Text", "Naming Rule"])

        return ZQRYFI_Eplant_without_Naming_Rule

    def ZECMM772_filter_out_duplicate(self, input_dataframe):
        input = input_dataframe.copy()
        DUPLCATE_LIST_unique = input[input[["Info Short Text", "Plant"]].duplicated()].drop_duplicates()

        dup_index = []
        for dup_PIR, dup_txt, dup_plt in zip(DUPLCATE_LIST_unique["PIR nr"], DUPLCATE_LIST_unique["Info Short Text"], DUPLCATE_LIST_unique["Plant"]):
            dup_slice = input_dataframe.loc[(input_dataframe["Info Short Text"]==dup_txt) & (input_dataframe["Plant"]==dup_plt)]
            if len(dup_slice) > 1 and len(dup_slice["PIR nr"].unique()) > 1:
                dup_index.append(dup_slice.index)

        dup_index_list = list(itertools.chain.from_iterable(dup_index))
        return input_dataframe.loc[~input_dataframe.index.isin(dup_index_list)], input_dataframe.loc[input_dataframe.index.isin(dup_index_list)]
    

    def main_process_tblRequestedtoCOE(self):
        diff_exist, diff_not_exist = [], []
        try:
            self.tblRequestedtoCOE["OAItemNumber_AGRP"].astype(float)
        except:
            box = tk.Tk()
            box.withdraw()
            messagebox.showinfo('Notification', "Unexpected data type of OAItemNumber_AGRP, Please check the tblRequestedtoCOE file and restart the GUI again!")
            
        self.tblRequestedtoCOE["Naming Rule"] = self.tblRequestedtoCOE["Eplant_name"].map(self.Source_file_plant_naming_mapping)
        self.tblRequestedtoCOE_Eplant_without_Naming_Rule = self.tblRequestedtoCOE[self.tblRequestedtoCOE["Naming Rule"].isna()]

        if len(self.tblRequestedtoCOE[self.tblRequestedtoCOE["Naming Rule"].isna()]["Eplant_name"]) !=0:
            print("tblRequestedtoCOE_Eplant without Naming Rule: ", self.tblRequestedtoCOE[self.tblRequestedtoCOE["Naming Rule"].isna()]["Eplant_name"].values)
        else:
            print("tblRequestedtoCOE_Eplant without Naming Rule: {}".format("N/A"))        
        
        self.tblRequestedtoCOE.loc[self.tblRequestedtoCOE['Level']!="PLC", "ID_KEY"] = self.tblRequestedtoCOE.loc[self.tblRequestedtoCOE['Level'] !="PLC"].apply(lambda x : self.NamingRule_processing(x, "Header_12NC", "Component_12NC", "Process_Text", "Eplant_name", diff_exist, diff_not_exist, self.REP_D077_drop_dup_map), axis=1)
        self.tblRequestedtoCOE.loc[self.tblRequestedtoCOE['Level']=="PLC", "ID_KEY"] = self.tblRequestedtoCOE.loc[self.tblRequestedtoCOE['Level']=="PLC", "Process_Text"] + " " + self.tblRequestedtoCOE.loc[self.tblRequestedtoCOE['Level']=="PLC", "Eplant_name"]


        print("Total Records in tblRequestedtoCOE: ", len(self.tblRequestedtoCOE))
        print("Total invalid in tblRequestedtoCOE: ", len(self.tblRequestedtoCOE.loc[self.tblRequestedtoCOE["ID_KEY"].str.contains("nan")]))
        print("Header invalid in tblRequestedtoCOE: ", len(self.tblRequestedtoCOE.loc[self.tblRequestedtoCOE["Header_12NC"].isna()]))
        print("Component invalid in tblRequestedtoCOE: ", len(self.tblRequestedtoCOE.loc[self.tblRequestedtoCOE["Component_12NC"].isna()]))
        return self.tblRequestedtoCOE_Eplant_without_Naming_Rule
    
    def main_process_PLC_PIR_per_Wave(self):
        diff_exist, diff_not_exist, REP_D077_drop_dup_map = [], [], []
        self.PLC_PIR_per_wave["Production Version"] = self.PLC_PIR_per_wave["Production Version"].fillna("")
        self.PLC_PIR_per_wave["Process Text"] = self.PLC_PIR_per_wave["Production Version"].apply(lambda x: self.Process_Text_processing(x))

        self.PLC_PIR_per_wave["Naming Rule"] = self.PLC_PIR_per_wave["Plnt"].map(self.Source_file_plant_naming_mapping)
        self.PLC_PIR_per_wave["ID_KEY"] =self.PLC_PIR_per_wave.apply(lambda x : self.NamingRule_processing(x, "Material Number", "BOM Component", "Process Text", "Plnt",
                                                                                        diff_exist, diff_not_exist, self.REP_D077_drop_dup_map), axis=1)
        pass

    def Process_Text_processing(self, input_string):
        output_str = ""
        if input_string == "PA+FT+PK>ASE-KHT,AS>ASE-KH":
            return "PA+AS+FT+PK"
        for idx, x in enumerate(input_string.split(",")):
            if len(input_string.split(",")) != 1:  #check if there is any process text contains ",", if yes, aggregate them and added "+" betweem actions
                output_str += x.split(">")[0]
                if idx != len(input_string.split(",")) - 1 :
                    output_str += "+"
            else:
                output_str += x.split(">")[0]
                continue
        return output_str
    
    def NamingRule_processing(self, input_slice, header, component, process_text, plant, diff_exist, diff_not_exist, REP_D077_drop_dup_map):
        Input12nc, Output12nc, Phase, Diff = "", "", "", ""
        output_str = ""
        input_str = str(input_slice["Naming Rule"])
        input_slice[header] = str(input_slice[header]).replace(" ","")
        input_slice[component] = str(input_slice[component]).replace(" ","")
        input_slice[header] = str(input_slice[header]).split(".")[0]
        input_slice[component] = str(input_slice[component]).split(".")[0]

        if "Input12nc" in input_str:
            Input12nc = input_slice[component]
            output_str += str(Input12nc) + " "        

        if "Output12nc" in input_str:
            Output12nc = input_slice[header]
            output_str += str(Output12nc) + " "
    
        if "Phase" in input_str:
            Phase = input_slice[process_text]
            output_str += str(Phase) + " "

        if "Diffusion name" in input_str:
            # REP_77.DWFR_Name = header 12NC
            # REP_77.DIFF_Name = Diffusion
            input_slice[header] = str(input_slice[header])
            try:
                Diff = REP_D077_drop_dup_map[input_slice[header]]
                output_str += Diff + " "
                diff_exist.append([input_slice[header], input_str, Diff])
            except:
                diff_not_exist.append([input_slice[component], input_slice[header], input_slice[process_text], input_str])
            pass
        
        output_str += input_slice[plant]
        return output_str

    def report_mapping(self, main_table, join_table, key_main_table, key_join_table, target_attr, target_naming, main_check_match_column, name_final_check_match_column, PLC_PIR_per):
        # ##################       Remark     ##################
        # #main_table = table to be joined, key_main_table = key to map to tables , target_attr = attrbiutes to be joined in, target_naming = the name of join attribute to be renamed
        # #join_table = join reference, key_join_table = key to map to tables
        # #main_check_match_column = the colmun in main table to be final compared, 
        # #name_final_check_match_column = check if the value in main main_check_match_column (oringal attributes in main table) is the same as the result of target_naming
        # PLC_PIR_per = if the input reference is PLC_PIR_Wave

        if PLC_PIR_per == True:
            id_mapping_dict= set(main_table.loc[main_table[name_final_check_match_column]=="Unmappable"][key_main_table]).intersection(join_table[key_join_table])
        else:
            id_mapping_dict= set(main_table[key_main_table]).intersection(join_table[key_join_table])

        id_mapping_target_dict = dict(zip(join_table.loc[join_table[key_join_table].isin(id_mapping_dict), key_join_table], 
                            join_table.loc[join_table[key_join_table].isin(id_mapping_dict), target_attr]))
        main_table.loc[main_table[key_main_table].isin(id_mapping_dict), target_naming] = main_table.loc[main_table[key_main_table].isin(id_mapping_dict), key_main_table].map(id_mapping_target_dict)
        
        #Additional condition for DI E028. E212, E312)
        if target_naming == "Cost Elem_SOURCE":
            main_table.loc[(main_table[key_main_table]=="DI") & ( ~main_table["Plnt"].isin(["E028", "E212", "E312"])), target_naming]=7701154
            main_table.loc[(main_table[key_main_table]=="DI") & (main_table["Plnt"]=="E028"), target_naming]=7701155
            main_table.loc[(main_table[key_main_table]=="DI") & (main_table["Plnt"]=="E212"), target_naming]=7734143
            main_table.loc[(main_table[key_main_table]=="DI") & (main_table["Plnt"]=="E312"), target_naming]=7734143

        # Check if the final mapping result matched current record in SAP
        main_table[name_final_check_match_column] = "N"
        main_table.loc[main_table[target_naming] == main_table[main_check_match_column], name_final_check_match_column] = "Y"
        main_table.loc[main_table[target_naming].isna(), name_final_check_match_column] = "Unmappable"
        return main_table

    def find_string(self, row, string):
        return any([string in str(cell) for cell in row])
    
    def cols_to_be_correct(self, input_slice, input_check_columns):
        output_stack = []
        for col in input_check_columns:
            if input_slice[col]=="N":
                output_stack.append(col)
        return output_stack

    #Check if there is more than one Process Text for each header 12nc under the same Eplant
    def broken_flow_check(self, input_dataframe, header_col, PV_text_col, Eplant_col):
        input_dataframe.loc[~input_dataframe[[header_col, Eplant_col]].duplicated(), "Broken_Flow"] = "N"
        DUPLCATE_LIST = input_dataframe.loc[input_dataframe[[header_col, Eplant_col]].duplicated(), [header_col, Eplant_col]]
        for dup_h, dup_p in zip(DUPLCATE_LIST[header_col], DUPLCATE_LIST[Eplant_col]):
            if len(input_dataframe.loc[(input_dataframe[header_col]==dup_h) & (input_dataframe[Eplant_col]==dup_p), PV_text_col].unique()) == 1:
                input_dataframe.loc[(input_dataframe[header_col]==dup_h) & (input_dataframe[Eplant_col]==dup_p), "Broken_Flow"] = "N"
            else:
                input_dataframe.loc[(input_dataframe[header_col]==dup_h) & (input_dataframe[Eplant_col]==dup_p), "Broken_Flow"] = "Y"
        return input_dataframe
```

```python
class MainWindow(QtWidgets.QMainWindow):
    def __init__(self):
        super(MainWindow, self).__init__() # Call the inherited classes __init__ method
        self.ui = Ui_MainWindow()
        self.ui.setupUi(self)

        self.today = (datetime.today()).strftime('%Y%m%d')
        self.Output_folder = '{cwd}/Output/{today_date}/'.format(cwd = os.getcwd().replace("\\", "/"), today_date = self.today)

        self.ICON_dict = {
                    "Download_ZQRYFI":{"LABEL":self.ui.Download_Label_ZQRYFI},
                    "Download_ZECMM772":{"LABEL":self.ui.Download_Label_ZECMM772},
                    "Download_REP077":{"LABEL":self.ui.Download_Label_REP077},
                    "Refresh_tblRequestedtoCOE":{"LABEL":self.ui.Refresh_Label_tblRequestedtoCOE},
                    "Refresh_PLC_Wave":{"LABEL":self.ui.Refresh_Label_PLC_Wave},
                    "Refresh_Source_Process_Text":{"LABEL":self.ui.Refresh_Label_Source_file_Process_Text},
                    "Refresh_Source_COEL_LORE":{"LABEL":self.ui.Refresh_Label_Source_file_COEL_LORE},
                    "Generate_ZQRYFI":{"LABEL":self.ui.Generate_Label_ZQRYFI_Consolidate},
                    "Generate_C223_LCO":{"LABEL":self.ui.Generate_Label_LCO_C223},
                    "Update_C223":{"LABEL":self.ui.SAP_Label_C223},
                    "Update_LCO":{"LABEL":self.ui.SAP_Label_LCO},
                    "Rename_Consolidate":{"LABEL":self.ui.Rename_End_Task},
                    "Upload_Sharepoint":{"LABEL":self.ui.Upload_SP_Label}
                    }
        
        # palette = QPalette()
        # self.setPalette(palette)



        # palette = QPalette()
        # palette.setBrush(QPalette.Background, QBrush(QPixmap("./Input/Background.png")))  
        # self.setPalette(palette)

        self.show()
        self.setup_control()
        self.setFixedSize(1371, 791) #fix window size

    def setup_control(self):
        self.ui.listView.setStyleSheet("QListView{background-image: url('./Input/background/Template_1.png')}")

        #Download
        self.ui.Download_Reports_All.clicked.connect(lambda : [self.Download_Reports_all(), self.pop_out_message("Process Done !!!")])
        self.ui.Download_ZQRYFI_TABLE.clicked.connect(lambda : [SAP_Download().ZQRYFI_Download("ZQRYFI"), self.try_button_work("Download_ZQRYFI")])
        self.ui.Download_ZECMM772_TABLE.clicked.connect(lambda : [SAP_Download().ZECMM772_Download("ZECMM772"), self.try_button_work("Download_ZECMM772")])
        self.ui.Download_REP077_TABLE.clicked.connect(lambda : [Download_ENOVIA_Reports().Download_REP_D077(), self.try_button_work("Download_REP077")])

        self.ui.Refresh_tblRequestedtoCOE.clicked.connect(lambda : [Sharepoint_Tables().Refresh_tblRequestedtoCOE(), self.try_button_work("Refresh_tblRequestedtoCOE")])
        self.ui.Refresh_PLC_Wave.clicked.connect(lambda : [Sharepoint_Tables().Refresh_PLC_PIR_per_wave(), self.try_button_work("Refresh_PLC_Wave")])
        self.ui.Refresh_Source_file_Process_Text.clicked.connect(lambda : [Sharepoint_Tables().Refresh_Source_file(), self.try_button_work("Refresh_Source_Process_Text")])
        self.ui.Refresh_Source_file_COEL_LORE.clicked.connect(lambda : [Sharepoint_Tables().Refresh_Source_file(), self.try_button_work("Refresh_Source_COEL_LORE")])

        #Data Preprocessing
        self.ui.Generate_ALL_Consolidate.clicked.connect(lambda : [self.Generate_Reports_all(), self.pop_out_message("Process Done !!!")])
        self.ui.Generate_ZQRYFI_Consolidate.clicked.connect(lambda : [Data_Preprocessing().main_consolidate_result(), self.try_button_work("Generate_ZQRYFI")])
        self.ui.Generate_C223_LCO.clicked.connect(lambda : [Data_Update().Generate_Files_to_be_Update(), self.try_button_work("Generate_C223_LCO")])

        #SAP Update
        self.ui.Update_via_C223.clicked.connect(lambda : [self.SAP_Update_Script_GUI("C223"), self.pop_out_message("Done with C223 Update!!!")])
        self.ui.Update_via_LCO.clicked.connect(lambda : [self.SAP_Update_Script_GUI("LCO"), self.pop_out_message("Done with LCO Update!!!")])

        self.ui.Button_Manual_Start.clicked.connect(lambda : [self.Manual_record_timer("Start")])
        self.ui.Button_Manual_End.clicked.connect(lambda : [self.Manual_record_timer("End")])

        #Check Button
        self.ui.Check_Generate_ZQRYFI_Consolidate.clicked.connect(lambda : os.startfile(self.Output_folder + "ZQRYFI_Consolidate_Result_{}.xlsx".format(self.today)))
        self.ui.Check_Generate_C223.clicked.connect(lambda : os.startfile(self.Output_folder + "C223.xlsx"))
        self.ui.Check_Generate_LCO.clicked.connect(lambda : os.startfile(self.Output_folder + "LCO.xlsx"))

        self.ui.Check_Update_Result.clicked.connect(lambda : os.startfile(self.Output_folder + "C223_LCO_Update_Consolidate_{}.xlsx".format(self.today)))
        self.ui.Check_Output_Folder.clicked.connect(lambda : os.startfile(self.Output_folder))
        self.ui.Check_Input_Folder.clicked.connect(lambda : os.startfile(self.Output_folder.replace("Output/{}/".format(self.today), "Input")))
        
        #Rename Consolidate
        self.ui.End_Task.clicked.connect(lambda : [self.Rename_consolidate_result(["ZQRYFI_Consolidate_Result",
                                                                                   "C223_LCO_Update_Consolidate"]), self.try_button_work("Rename_Consolidate"), self.pop_out_message("Action Done!!!")])
        # Result Upload
        self.ui.Upload_SP_button.clicked.connect(self.Upload_final_result)
        #Stop
        self.ui.Stop_GUI_button.clicked.connect(self.CLOSED_GUI)


    def SAP_Update_Script_GUI(self, update_type):
        Update_start_time = time.time()
        if update_type == "C223":
            try:
                output_frame, Output_header = Data_Update().SAP_Update(update_type)
                Data_Update().Export_C223_LCO(output_frame, update_type, Output_header)
                self.ICON_dict["Update_C223"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")

                Time_spent = round((time.time()-Update_start_time)/60, 2)
                self.ui.Display_Update_Time_C223.setText("{} mins".format(Time_spent))
                print("Total time spent for {type} update: {time_spent} mins! ".format(type = update_type, time_spent = Time_spent))

            except:
                self.ICON_dict["Update_C223"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")

        if update_type == "LCO":
            try:
                output_frame, Output_header = Data_Update().SAP_Update(update_type)
                Data_Update().Export_C223_LCO(output_frame, update_type, Output_header)
                self.ICON_dict["Update_LCO"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
                
                Time_spent = round((time.time()-Update_start_time)/60, 2)
                self.ui.Display_Update_Time_LCO.setText("{} mins".format(Time_spent))
                print("Total time spent for {type} update: {time_spent} mins! ".format(type = update_type, time_spent = Time_spent))

            except:
                self.ICON_dict["Update_LCO"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")             
        pass


    def Download_Reports_all(self):
        Download_start_time =time.time()
        [os.remove(x) for x in glob.glob('./Input/*.xlsx')]
        [os.remove(x) for x in glob.glob('./Input/*.gz')]

        #ZQRYFI
        try:
            SAP_Download().ZQRYFI_Download("ZQRYFI")
            self.ICON_dict["Download_ZQRYFI"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ICON_dict["Download_ZQRYFI"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")

        #ZECMM772
        try:
            SAP_Download().ZECMM772_Download("ZECMM772")
            self.ICON_dict["Download_ZECMM772"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ICON_dict["Download_ZECMM772"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")
        
        #Download_REP_D077
        try:
            Download_ENOVIA_Reports().Download_REP_D077()
            self.ICON_dict["Download_REP077"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ICON_dict["Download_REP077"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")
        
        #Import_tblRequestedtoCOE
        try:
            Sharepoint_Tables().Refresh_tblRequestedtoCOE()
            self.ICON_dict["Refresh_tblRequestedtoCOE"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ICON_dict["Refresh_tblRequestedtoCOE"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")

        #Import_PLC_PIR_per_wave
        try:
            Sharepoint_Tables().Refresh_PLC_PIR_per_wave()
            self.ICON_dict["Refresh_PLC_Wave"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ICON_dict["Refresh_PLC_Wave"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")

        #Import_Source_file
        try:
            Sharepoint_Tables().Refresh_Source_file()
            self.ui.Refresh_Label_Source_file_Process_Text.setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
            self.ui.Refresh_Label_Source_file_COEL_LORE.setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ui.Refresh_Label_Source_file_Process_Text.setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")
            self.ui.Refresh_Label_Source_file_COEL_LORE.setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")
    
        print("\nTotal time spent for data downloading: {time_spent} mins! \n".format(time_spent = round((time.time()-Download_start_time)/60, 2)))
        self.ui.Display_Download_Time.setText("{} mins".format(round((time.time()-Download_start_time)/60, 2)))
        pass

    def Generate_Reports_all(self):
        Generation_start_time =time.time()
        try:
            Data_Preprocessing().main_consolidate_result()
            self.ICON_dict["Generate_ZQRYFI"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ICON_dict["Generate_ZQRYFI"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")

        try:
            Data_Update().Generate_Files_to_be_Update()
            self.ICON_dict["Generate_C223_LCO"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ICON_dict["Generate_C223_LCO"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")
    
        print("\nTotal time spent for report generating: {time_spent} mins! \n".format(time_spent = round((time.time()-Generation_start_time)/60, 2)))
        self.ui.Display_Processing_Time.setText("{} mins".format(round((time.time()-Generation_start_time)/60, 2)))
        pass

    def Rename_consolidate_result(self, input_file_name_to_be_convert):
        output_datetime = (datetime.now()).strftime("%Y%m%d_%H%M")
        for f in input_file_name_to_be_convert:
            os.rename("./Output/{date}/{file_name}_{date}.xlsx".format(date = self.today, file_name = f),
                      "./Output/{date}/{file_name}_{datetime}.xlsx".format(date = self.today,
                                                                             file_name = f,
                                                                             datetime = output_datetime,
                                                                             ))
            print("{} is now renamed !".format(f))
        pass


    def Upload_final_result(self):
        subprocess.call(["taskkill", "/f", "/im", "EXCEL.EXE"])
        time.sleep(1)
        src = "./Output/{date}/".format(date = self.today)
        dst = "./Buffer/{date}/".format(date = self.today)
        if not os.path.exists(dst): 
                os.makedirs(dst)
                
        try:
            for x in os.listdir(src):
                shutil.copy(src + x, dst)

            self.ICON_dict["Upload_Sharepoint"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
            self.pop_out_message("Folder has been uploaded successfully!!!")
            os.startfile(os.getcwd().replace("\\", "/") + dst.replace(".", ""))
            
        except:
            self.ICON_dict["Upload_Sharepoint"]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")
            self.pop_out_message("Folder was uploaded unsuccessfully, please check if the source files/folder are correctly maintained!!!")
            os.startfile(os.getcwd().replace("\\", "/") + src.replace(".", ""))
            pass        

    def try_button_work(self, input_icon_lable_name):
        try:
            self.ICON_dict[input_icon_lable_name]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Sign-check-icon.png)")
        except:
            self.ICON_dict[input_icon_lable_name]["LABEL"].setStyleSheet("border: rgba(0, 0, 0, 0); image : url(./Input/Icon/Error-icon.png)")
        pass

    def Manual_record_timer(self, input_type):
        if input_type == "Start":
            global Manual_start_time
            Manual_start_time = time.time()
            self.ui.Display_Manual_Start.setText(time.strftime("%Y/%m/%d %H:%M:%S"))
            self.ui.Display_Manual_End.setText("")
            
        if input_type == "End":
            Manual_end_time = time.time()
            self.ui.Display_Manual_End.setText(time.strftime("%Y/%m/%d %H:%M:%S"))
            self.ui.Display_Manual_Process.setText("{} mins".format(round((Manual_end_time - Manual_start_time)/60, 2)))
    
    def pop_out_message(self, input_string):
        box = tk.Tk()
        box.withdraw()
        messagebox.showinfo('Notification', input_string)
        pass

    def CLOSED_GUI(self):
        self.close()
        #sys.exit(app.exec_())

```
